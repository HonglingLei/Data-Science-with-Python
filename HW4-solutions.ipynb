```python
import pandas as pd
import numpy as np
import time
import os
os.getcwd()
```

```python
file = "/Users/hongling/Desktop/HW4/pensioners.csv"
data = pd.read_csv(file)
```

```python
class aggregator(object):

    def __init__(self):
        self.xsum=0.0;
        self.n=0.0
        
    def add(self, x):
        self.xsum+=x
        self.n+=1
        
    def count(self):
        return self.n
    
    def sum(self):
        return self.xsum
    
    def mean(self):
        self.mean=self.xsum/self.n
        return self.mean
```

```python
# 新建一个空表来装annual indicators
indicators=pd.DataFrame(index=list(range(2015,2041)),columns=["Total pension","Total pension of all males","Total pension of all females","Average age of all pensioners","Number of all pensioners","Average pension amount","Average pension amount (male)","Average pension amount (female)"])
```

```python
data.loc["813691"]=0 #加上一行，不然下面的代码总是报错“single positional indexer is out-of-bounds”。。。。
data.iloc[[-1],[0]]=2041
```

```python
t0=time.time()

pension=aggregator()
age=aggregator()
male_pension=aggregator()
female_pension=aggregator()
year=2015
i=0

while data.shape[0]-i: #从第一行到最后一行
    if data.iloc[i][0]==year : #对当前年份
        pension.add(data.iloc[i][2]) #把这一年所有lastpension加起来
        age.add(data.iloc[i][3]) #把这一年所有age加起来
        if data.iloc[i][5]==1: #如果为男性
            male_pension.add(data.iloc[i][2]) #就加到male_pension里面去
        else: #如果为女性
            female_pension.add(data.iloc[i][2]) #就加到female_pension里面去
        i+=1 #转到下一行
    else: #如果year到了下一年
        indicators.loc[year]["Total pension"]=pension.sum()
        indicators.loc[year]["Total pension of all males"]=male_pension.sum()
        indicators.loc[year]["Total pension of all females"]=female_pension.sum()
        indicators.loc[year]["Average age of all pensioners"]=age.mean()
        indicators.loc[year]["Number of all pensioners"]=pension.count()
        indicators.loc[year]["Average pension amount"]=pension.mean()
        indicators.loc[year]["Average pension amount (male)"]=male_pension.mean()
        indicators.loc[year]["Average pension amount (female)"]=female_pension.mean()
        year+=1
        pension=aggregator()
        age=aggregator()
        male_pension=aggregator()
        female_pension=aggregator()
        
print(time.time()-t0) #execution time
```
